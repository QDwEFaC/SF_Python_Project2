Для начала обсудим условия решения задачи:

* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
# функция извлечения опыта работы в месяцах
def get_work_exp(text: str) -> int:
    # nan - оставляем
    if text is np.nan:
        return np.nan
    
    # не указано - преобразуем в nan
    if text.lower() == 'не указано':
        return np.nan
    
    # если в тексте меньше 3-х полей - такого не должно быть, заполняем как nan
    parts = text.split(' ')
    if len(parts) < 3:
        return np.nan
    
    # инициализируем переменные
    years = 0
    month = 0

    # итерируемся по блокам текста
    for i in range(6):
        if len(parts) <= i:
            break

        # если в блоке есть слово год - это год
        if 'год' in parts[i] or 'лет' in parts[i]:
            try:
                years = int(parts[i-1])
            except ValueError:
                print(f'Ошибка парсинга года: входная строка: {text}, поле год: {parts[i-1]}')

        # если в блоке есть слово месяц - это месяцы
        if 'месяц' in parts[i]:
            try:
                month = int(parts[i-1])
            except ValueError:
                print(f'Ошибка парсинга месяца: входная строка: {text}, поле месяц: {parts[i-1]}')
            break

    return years * 12 + month

# заполняем новый признак
data_df['Опыт работы (месяц)'] = data_df['Опыт работы'].apply(get_work_exp)
# удаляем старый признак
data_df.drop('Опыт работы', axis=1, inplace=True)

# считаем значение для задания
med_work = data_df['Опыт работы (месяц)'].median()
print(f'Медианный опыт работы = {med_work}')
```

Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
# функция для разделения признаков
def split_categories(col_name, data_df):
    # формируем список новых признаков
    target_cat = []
    for var in data_df[col_name]:
        parts = var.split(',')
        for x in parts:
            cat = x.strip()
            if cat not in target_cat:
                target_cat.append(cat)
    print(target_cat)

    # функция извлечения нового признака из старого
    def extract(text):
        res = [False] * len(target_cat)
        for i, cat in enumerate(target_cat):
            if cat in text:
                res[i] = True
        return res
    
    # извлекаем новые признаки
    new_df = data_df.copy()
    new_df[target_cat] = new_df[col_name].apply(lambda x: pd.Series(extract(x)))
    return new_df


# заполняем новые признаки
split_data_df = split_categories('Занятость', data_df)
split_data_df = split_categories('График', split_data_df)
# удаляем старый признак
split_data_df.drop(['Занятость', 'График'], axis=1, inplace=True)
```

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь
# загружаем dataset с курсом валют
columns_to_load = ['currency', 'date', 'proportion', 'close']
rates = pd.read_csv('./data/ExchangeRates.csv', header=0, delimiter=',', usecols=columns_to_load)

# преобразуем дату
rates['date'] = pd.to_datetime(rates['date'])
rates['date'] = rates['date'].apply(lambda x: x.date())
display(rates.head())
rates.info()
```

```python
# функция ковертации валюты
def convert(name, value, date):
    global rates

    if name == 'руб.':
        return value
    
    raw_names = {
        'грн.': 'UAH', 
        'USD': 'USD', 
        'EUR': 'EUR', 
        'бел.руб.': 'BYN', 
        'KGS': 'KGS', 
        'сум': 'UZS', 
        'AZN': 'AZN', 
        'KZT': 'KZT'
    }

    # проверка, что такая валюта есть в базе
    if name not in raw_names:
        raise Exception(f'Currency {name} not found in rates list!')

    # извлекаем курс валюты п дате и названию
    mask = (rates['currency'] == raw_names[name]) & (rates['date'] == date)
    entry = rates[mask]
    if entry.shape[0] == 0:
        raise Exception(f'No entry for {raw_names[name]} and {date}')
    # вычисляем конвертацию
    res = (entry['close'].iloc[0] / entry['proportion'].iloc[0]) * value
    return res

# функция формирования нового признака
def create_pay(curr_pay, up_date):
    try:
        date = pd.to_datetime(up_date, format='%d.%m.%Y %H:%M').date()
    except Exception as e:
        print(f'Parsing error for date: {up_date}')
    parts = curr_pay.split(' ')
    num = int(parts[0].strip())
    currency = parts[1].strip()
    try:
        res = convert(currency, num, date)
    except Exception as e:
        print(f'Error in convert for input {curr_pay} {up_date}')
    return res

# формируем новый признак
split_data_df['ЗП (руб)'] = split_data_df.apply(lambda row: create_pay(row['ЗП'], row['Обновление резюме']), axis=1)
# удаляем старый признак
split_data_df.drop('ЗП', axis=1, inplace=True)
# значение для решения задания
print('Медианная зарплата: ', split_data_df['ЗП (руб)'].median())
```

```python
# ваш код здесь
# функция для базового анализа
def analyze_distribution(df, col):
    
    # Основные статистики
    print("ОСНОВНЫЕ СТАТИСТИКИ:")
    print(df[col].describe())
    
    # Мода распределения
    mode_value = df[col].mode()
    print(f"\nМОДА РАСПРЕДЕЛЕНИЯ: {mode_value.values[0]}")
    print(f"   (наиболее часто встречающееся значение)")
    
    # Предельные значения
    min_val = df[col].min()
    max_val = df[col].max()
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    print(f"\nПРЕДЕЛЬНЫЕ ЗНАЧЕНИЯ:")
    print(f"   Минимум: {min_val}")
    print(f"   Максимум: {max_val}")
    
    print(f"   Большинство показателей находится в интервале: {Q1:.1f} - {Q3:.1f}")
    return Q1, Q3, IQR

# Визуализация распределения
def plot_distribution(df, column):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
    histplot = sns.histplot(data=pd.DataFrame(df[column]), x=column, ax=axes[0]);
    histplot.set_title(f'Распределение {column}');
    sns.boxplot(x=df[column])
    plt.title('Boxplot')
    plt.show()

# метод сигм
def outliers_iqr_mod(data, feature, log_scale=False, left=1.5, right=1.5):
    data_df = data.copy()
    x = data_df[feature]
    if log_scale:
        x = np.log(x)
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75),
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * left)
    upper_bound = quartile_3 + (iqr * right)
    outliers = data_df[(x < lower_bound) | (x > upper_bound)]
    cleaned = data_df[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

# метод z-score
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

# Анализ аномалий
def detect_anomalies(df: pd.DataFrame, column):
    print("\nАНАЛИЗ АНОМАЛИЙ:")
    
    print(f'Стандартное отклонение: {round(df[column].std(), 2)}')
    print(f'Вариация данных (CV): {round((df[column].std() / df[column].mean()) * 100, 2)}')

    outliers, cleaned = outliers_iqr_mod(df, column)
    age_out = outliers[column].unique()
    age_out_sorted = sorted(age_out)
    s = ""
    if len(age_out_sorted) > 10:
        s = f'{age_out_sorted[0]}, {age_out_sorted[1]}, {age_out_sorted[2]} ... {age_out_sorted[-3]}, {age_out_sorted[-2]}, {age_out_sorted[-1]}'
    else:
        for i, x in enumerate(age_out_sorted):
            if (i % 10) == 0:
                s += '\n\t'
            s += f'{x} '
    print(f'Выбросы по методу Тьюки (всего {age_out.shape[0]}): {s}')

    outliers, cleaned = outliers_z_score_mod(df, column)
    age_out_z = outliers[column].unique()
    age_out_z_sorted = sorted(age_out_z)
    s = ""
    if len(age_out_z_sorted) > 10:
        s = f'{age_out_z_sorted[0]}, {age_out_z_sorted[1]}, {age_out_z_sorted[2]} ... {age_out_z_sorted[-3]}, {age_out_z_sorted[-2]}, {age_out_z_sorted[-1]}'
    else:
        for i, x in enumerate(age_out_z_sorted):
            if (i % 10) == 0:
                s += '\n\t'
            s += f'{x} '
    print(f'Выбросы по методу сигм (всего {age_out_z.shape[0]}): {s}')
    return

age_column = 'Возраст'
# Проверяем на пропуски
missing_count = split_data_df[age_column].isna().sum()
print(f"Пропущенные значения: {missing_count}")

# Базовый анализ
Q1, Q3, IQR = analyze_distribution(split_data_df, age_column)

# Анализ аномалий
detect_anomalies(split_data_df, age_column)

print(f"   Соискатели младше 18 лет: {len(split_data_df[split_data_df[age_column] < 18])}")
print(f"   Соискатели старше 65 лет: {len(split_data_df[split_data_df[age_column] > 65])}")

# Визуализация
plot_distribution(split_data_df, age_column)

print(f"Большинство показателей находятся в диапазоне от {Q1:.0f} до {Q3:.0f}")
```

```python
# ваш код здесь
# Создаем обощённые возрастные группы
bins_detailed = [18, 22, 26, 30, 35, 40, 45, 50, 55, 60, 65, 75]
labels_detailed = ['18-22', '23-26', '27-30', '31-35', '36-40', '41-45', '46-50', '51-55', '56-60', '61-65', '66+']
filtered_df['Возрастная_группа_детально'] = pd.cut(filtered_df['Возраст'], bins=bins_detailed, labels=labels_detailed, right=False)

# Создаем сводную таблицу
pivot_detailed = filtered_df.pivot_table(
    values='ЗП (руб)',
    index='Образование',
    columns='Возрастная_группа_детально',
    aggfunc='median'
)

# Заполняем пропуски средним по строке
pivot_detailed = pivot_detailed.fillna(pivot_detailed.mean(axis=1), axis=0)

fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# 1. Тепловая карта с детальными возрастными группами
sns.heatmap(
    pivot_detailed,
    annot=True,
    fmt=',.0f',
    cmap='RdYlBu_r',
    linewidths=0.5,
    linecolor='white',
    ax=axes[0],
    cbar_kws={'label': 'Медианная зарплата (руб)'}
)
axes[0].set_title('Детальная тепловая карта\nОбразование vs Возраст', fontweight='bold')
axes[0].set_xlabel('Возрастная группа')
axes[0].set_ylabel('Образование')

# 2. Нормализованная тепловая карта (z-score по строкам)
pivot_normalized = pivot_detailed.apply(lambda x: (x - x.mean()) / x.std(), axis=1)

sns.heatmap(
    pivot_normalized,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    linewidths=0.5,
    linecolor='white',
    ax=axes[1],
    cbar_kws={'label': 'Z-score (отклонение от среднего)'}
)
axes[1].set_title('Нормализованная тепловая карта\n(относительно среднего по образованию)', fontweight='bold')
axes[1].set_xlabel('Возрастная группа')
axes[1].set_ylabel('Образование')

plt.tight_layout()
plt.show()
```

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# Создаем логарифмическую версию возраста (добавляем 1 чтобы избежать log(0))
#data_df['log_возраст'] = np.log1p(data_df['Возраст'])
data_df['log_возраст'] = np.log(data_df['Возраст'])

# Строим распределение в логарифмическом масштабе
plt.figure(figsize=(12, 6))

# Гистограмма в логарифмическом масштабе
histplot = sns.histplot(data=data_df, x='log_возраст', bins=30, kde=True)
plt.title('Распределение возраста в логарифмическом масштабе')
plt.xlabel('ln(Возраст + 1)')
plt.ylabel('Частота')

# Добавляем линии для среднего и границ ±3 сигм
log_age = data_df['log_возраст']
mean_log = log_age.mean()
std_log = log_age.std()

histplot.axvline(mean_log, color='red', lw=2, label=f'Среднее: {mean_log:.2f}')
histplot.axvline(mean_log + 3*std_log, color='orange', lw=2, linestyle='--', 
                label=f'+3*sigma: {mean_log + 3*std_log:.2f}')
histplot.axvline(mean_log - 3*std_log, color='orange', lw=2, linestyle='--', 
                label=f'-3*sigma: {mean_log - 3*std_log:.2f}')

plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Находим выбросы с помощью z-отклонения (левая сигма = 3, правая = 4)
z_scores = (log_age - mean_log) / std_log
left_outliers_mask = z_scores < -3  # 3 сигмы в левую сторону
right_outliers_mask = z_scores > 4   # 4 сигмы в правую сторону
total_outliers_mask = left_outliers_mask | right_outliers_mask

# Преобразуем обратно в обычный возраст для анализа
left_threshold_log = mean_log - 3 * std_log
right_threshold_log = mean_log + 4 * std_log
left_threshold = np.expm1(left_threshold_log)
right_threshold = np.expm1(right_threshold_log)

print(f"\nГРАНИЦЫ ДЛЯ ВЫБРОСОВ:")
print(f"Левая граница (3*sigma): {left_threshold:.1f} лет")
print(f"Правая граница (4*sigma): {right_threshold:.1f} лет")

print(f"\nНАЙДЕНО ВЫБРОСОВ:")
print(f"Всего выбросов: {total_outliers_mask.sum()}")
print(f"Выбросы слева (возраст < {left_threshold:.1f} лет): {left_outliers_mask.sum()}")
print(f"Выбросы справа (возраст > {right_threshold:.1f} лет): {right_outliers_mask.sum()}")

# Таблица с выбросами
if total_outliers_mask.sum() > 0:
    outliers_df = data_df[total_outliers_mask][['Возраст', 'Пол', 'Образование', 'Город', 'ЗП (руб)']].copy()
    outliers_df['Тип_выброса'] = np.where(
        data_df[total_outliers_mask]['Возраст'] < left_threshold, 
        'Слишком молодой', 
        'Слишком пожилой'
    )
    
    # Удаляем выбросы из данных
    initial_count = len(data_df)
    data_df = data_df[~total_outliers_mask]
    final_count = len(data_df)
    
    print(f"\nРЕЗУЛЬТАТ УДАЛЕНИЯ:")
    print(f"Удалено резюме: {initial_count - final_count}")
    print(f"Осталось резюме: {final_count}")
    
    # Удаляем временный столбец
    data_df = data_df.drop('log_возраст', axis=1)
    
else:
    print("Выбросы не найдены")
```
